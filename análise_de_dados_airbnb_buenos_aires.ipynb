<h2> História da cidade </h2>

<center> Reprodução:<a href="https://www.buenosaires.gob.ar/planeamiento/archivo-fotografico"> Site cidade de Buenos Aires</a> </center>

<p> A cidade de Buenos Aires passou por dois processos de fundação. O primeiro ocorreu em 1536, realizado por Pedro de Mendoza. O segundo foi realizado for Juan de Garay em 1580 quando este partira em uma expedição com 100 homens a partir de Assunção. Ao chegar, em 11 de Julho de 1580, estabeleceu a Ciudad de la Santíssima Trinidad y Puerto de Santa María del Buen Ayre.

A partir deste momento, começaram as distribuições de terras a muitos daqueles que o acompanharam, estabelecendo-se o Cabildo de Buenos Aires. A cidade era considerada um bom ponto estratégico uma vez que as águas rasas do Rio da Prata ofereciam aos novos habitantes uma defesa natural contra navios inimigos, os quais não conseguiam chegar diretamente ao continente.

No início do século XVII, Buenos Aires era composta por 250 blocos retangulares, com um forte, três conventos e várias casas de barro e palha. Em 1776, com a criação do Vice-Reino do Rio da Prata, Buenos Aires é convertida em capital e começa o seu desenvolvimento com o aumento da atividade comercial, uma vez que a cidade permitia uma comunicação fluida com a Espanha através do Atlântico.

Entre 1780 e 1800, a cidade progrediu substancialmente recebendo um forte fluxo de imigrantes de várias nacionalidades dentre eles espanhois, italianos, alemães e russos. Esse fluxo migratório deramdeu origem a uma cidade cosmopolita onde diferentes culturas e religiões convivem lado a lado.

<h3> Analisando as colunas <em><strong>price</em></strong> e <em><strong>minimum nights</em></strong></h3>

<h4> Coluna <em><strong>price</em></strong></h4>

<p> Vamos construir um diagrama de caixa. Esse gráfico é ideal para a identificação de outliers, pois os dados numéricos são dividos em quartis (que representam 25% dados) e formando uma caixa cujas extremidas superiores e inferiores são o primeiro e o terceiro quartis, respectivamente. Além disso, a mediana é assinalada com um linha entre as extemidades.

```python
#...Análise preço
fig, ax = plt.subplots( figsize=(12,6))
sns.boxplot(x='price',data=df,fliersize=5)
```

```python
#...Descrição estatística da coluna
df['price'].describe()
```

<p> O nosso gráfico nos mostra que os valores estão mal distribuídos. Uma estratégia para tratar os dados é regra do interquartil, onde os valores fora dos limites superior e inferior são considerados outliers. O cálculo desses limites é mostrado abaixo

Limite inferior: $$ Q_{1} - FIQ  $$
Limite superior: $$ Q_{3} + FIQ  $$
Faixa interquartil : $$ FIQ = Q_{3} - Q_{1}$$

onde:<br><br> $ Q_{1} $ é o primeiro quartil, $ Q_{3} $ é o terceiro quartil e $ FIQ $ é a faixa interquartil

```python
#...Aquisição dos valores dos quartis
q1 = df['price'].describe()[4]
q3 = df['price'].describe()[6]
fiq = q3 - q1

#...Cálculo dos limites inferiores e superiores
lb = q1 - fiq
up = q3 + fiq

#...Remoção de outliers
df2 = df[(df['price']>= lb)&(df['price']<=up)].copy()

print(f'Retirando os outliers, ficamos com {df2.shape[0]} entradas. Isso corresponde a {round(df2.shape[0]*100/df.shape[0],2)} % dos dados ')
```

```python
#...Novo diagrama de caixa
sns.boxplot(x='price',data=df2,fliersize=5)
```

<h4> Análise de <strong><em>minimum_nights</em></strong> </h4>

Assim como na análise anterior, iniciaremos com a construção de um diagrama de caixa

```python
#...Contrução do boxplot
sns.boxplot(x='minimum_nights',data=df2)
plt.show()
```

<p> Assim como os dados para o preço, a coluna de noites mínimas também se encontra com dados desbalanceados. Vamos usar a mesma regra dos interquartis para a remoção de outliers.

```python
#...Aquisição dos valores dos quartis
q1 = df2['minimum_nights'].describe()[4]
q3 = df2['minimum_nights'].describe()[6]
fiq = q3 - q1

#...Cálculo dos limites inferiores e superiores
lb = q1 - fiq
up = q3 + fiq

#...Remoção de outliers
df3 = df2[(df2['minimum_nights']>= lb)&(df2['minimum_nights']<=up)].copy()

print(f'Após a limpeza de dados, ficamos com {df3.shape[0]} entradas. Isso representa {df3.shape[0]*100/df2.shape[0]} % das entradas do conjunto de dados anterior.')
print('')
print(f'Ao final do processo de limpeza, ficamos com {df3.shape[0]*100/df.shape[0]} % dos dados originais, o que corresponde a {df3.shape[0]} entradas.')

```

```python
sns.boxplot(x='minimum_nights',data=df3)
plt.show()
```

<p> Recalculando a média de preços e noites mínimas para o dataset tratados, temos que:

```python
df3[['price','minimum_nights']].describe()
```

```python
#...Construção dos novos diagramas de caixa com a remoção de outliers de ambas colunas
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(12,8))

sns.boxplot(x='price',data=df3,ax=ax1)
sns.boxplot(x='minimum_nights',data=df3,ax=ax2)

ax1.set_xlabel('Preço (em pesos)')
ax2.set_xlabel('Número mínimo de noites')
plt.show()
```

<h3>Distribuição do preços médios</h3>

<p> Agora vamos avaliar a distribuição dos preços médios de alocação de imóveis pelos bairros a partir de um gráfico cloroplético (<em>choropleth</em>). Os gráficos coropléticos utilizam cores para mostrar dados normalizados como áreas,linhas ou pontos sombreados. 

<p> Para a construção desse gráfico, vamos seguir três etapas:
    <ol>
        <li> Verificação do nome dos bairros
        <li> Contrução de um novo DataFrame com a média de preços para cada bairro
        <li> Importar um arquivo geojson (um formato para codificar uma variedade de estrutura de dados geográficos) como um geoDataFrame da biblioteca geopandas
        <li> Depois vamos unir esse dois DataFrames para que a informações das etapas 2 e 3 estejam no mesmo geoDataFrame
        <li> Construção do mapa com a biblioteca folium. A biblioteca folium permite a vinculação de dados a um mapa para visualizações coropléticas e também a transmissão de visualizações ricas de vetor/raster/HTML como marcadores no mapa.
    </ol>

```python
#...Verificação dos bairros
dfclean['neighbourhood'].unique()
```

<p> Olhando na <a href="https://www.buenosaires.gob.ar/laciudad/barrios"> lista de bairros da cidade </a> , as entradas que contém a palavra <em> Dique </em>(Diques 1,2,3 e 4) não são bairros oficiais da cidade. Nesses anúncios, os proprietários listam os imóveis em Puerto Madero. Portanto, esse bairros serão incluídos no bairro de Puerto Madero 

```python
#...Anuncios listados como sendo do bairro dique
diques = ['Dique']
dfclean[dfclean['neighbourhood'].str.contains(diques[0])]
```

```python
#...Consertando os bairros diques
dfclean['neighbourhood'] = dfclean['neighbourhood'].replace({'Dique 1':'Puerto Madero','Dique 2':'Puerto Madero',
                                'Dique 3':'Puerto Madero','Dique 4':'Puerto Madero'})    

dfclean['neighbourhood'].unique()
```

```python
#...Preparação dos dados

#...Construção do dataframe com a quantidade de imóveis disponíveis para cada bairro
df11 = dfclean['neighbourhood'].value_counts()
df11 = pd.DataFrame(df11).reset_index()
df11.rename({'index': 'neighbourhood', 'neighbourhood':'number_listing'},axis=1,inplace=True)
df11.head()
```

```python
#...Construção do dataframe com a média de preços para cada bairro
df12 = dfclean.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
df12 = pd.DataFrame(df12).reset_index()
df12.head(10)
```

```python
#...União dos dois dataframes criados a cima
df13 = df11.merge(df12,how = 'left', on='neighbourhood')
df13.head()
```

```python
import geopandas as gpd

#...Importação do arquivo geojson
geojson = gpd.read_file('neighbourhoods.geojson')
geojson.head()
```

```python
#...União das informações em um único geodataframe
df13 = geojson.merge(df13,on='neighbourhood',how='left')
df13.head()
```

```python
import folium

arg_latitude = -34.603722
arg_longitude = -58.381592

#...Criação do mapa
m2 = folium.Map(location = [arg_latitude,arg_longitude],zoom_start=11)

#...Criação do mapa coroplético
folium.Choropleth(
            geo_data='neighbourhoods.geojson',
            data=df13,
            columns=['neighbourhood', 'price'],  
            key_on='feature.properties.neighbourhood',
            fill_color='YlOrRd',
            nan_fill_color="White", 
            fill_opacity=0.5,
            line_opacity=0.2,
            legend_name='Média de preços por bairro',
            highlight=True,
            line_color='black').add_to(m2) 

#...Adição de visualizações no gráfico
folium.features.GeoJson(
    data = df13,
    name = "Mapa",
    smooth_factor = 2,
    style_function=lambda x: {'color': 'black', 'fillColor':'transparent','weight':0.5},
    tooltip = folium.features.GeoJsonTooltip(
        fields = ['neighbourhood','number_listing','price'],
        aliases=['Bairro','Anúncios','Preço'],
        localize = True,
        sticky=False,
        labels=True,
        ),
    highlight_function=lambda x: {'weight':3,'fillColor':'grey'}
).add_to(m2)

m2
```

<p> O gráfico coroplético não nos diz diretamente quais os bairros que apresentam a maior oferta de imóveis, então vamos construir dois gráficos de barra comparando o preço médio de aluguel e o quantidade total de imóveis disponíveis.

```python
#...Construção dos gráficos
fig,(ax1,ax2)= plt.subplots(1,2,figsize=(12,8),sharey=True)

df14 = df13.copy()
df14.set_index('neighbourhood',inplace=True)


sns.barplot(x='price',y='neighbourhood',data=df13,ax=ax1) 
sns.barplot(x='number_listing',y='neighbourhood',data=df13,ax=ax2) 
ax1.set_xlabel('Preço médio por noite (em pesos argentinos)')
ax2.set_xlabel('Quantidade total de anúncios disponíveis')
ax1.set_ylabel('Bairros')
ax2.set_ylabel('')
plt.tight_layout()

plt.show()
```

<p> Como pode ser visto nos gráficos, Puerto Madero é o bairro com o maior preço médio de aluguel, entretanto oferece poucos imóveis. Enquanto isso, o bairro de Palermo apresenta o quinto maior preço e a maior disponibilidade de anúncios.
